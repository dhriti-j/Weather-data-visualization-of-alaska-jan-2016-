```python
alaska=df[df["Station.State"]=="Alaska"]
alaska
```

```python
alaska["Date.Month"].value_counts()
```

```python
alaska1= alaska[(alaska["Date.Month"]==1) & (alaska["Date.Year"]==2016)]
alaska1["Date.Full"].value_counts()
```

```python
alaska1["Station.Code"].value_counts()
```

```python
# We have data of 31 codes on all 5 dates of january 2016
```

```python
codes =[]
for loc in alaska1["Station.Code"]:
    if loc not in codes:
        codes.append(loc)
codes
```

```python
x=[]
y=[]
plt.figure(figsize = (20,8))
plot1=plt.scatter(x,y)
overlapping= 0.5
colors = plt.cm.hsv(np.linspace(0, 1,32))
def plotting(code,colour):
    date=alaska1[alaska1["Station.Code"]==code]
    plot1=plt.plot(date["Date.Full"],date["Data.Temperature.Avg Temp"],marker="o", label = code, color=colour,alpha=overlapping)
    plt.legend(fontsize=7, loc="upper right")


for i,loc in enumerate(codes):
    colour= colors[i]
    plotting(loc,colour)
   



plt.xlabel('Dates')
plt.ylabel('Average Temperature(F)')
plt.title('Average Temperature of each station code')



    

 

```

```python
new= {"Code": [],
     "Min": [],
     "Max": [],
     "Avg": []}
new=pd.DataFrame(new)
   
    
for loc in codes:
     alaska_loc = alaska1[alaska1["Station.Code"] == loc]
     
     max_temp= alaska_loc["Data.Temperature.Max Temp"].max()
     min_temp= alaska_loc["Data.Temperature.Min Temp"].min()
     avg_temp=0
     for temp in alaska_loc["Data.Temperature.Avg Temp"]:
       avg_temp= avg_temp + temp
     avg_temp= avg_temp/len(alaska_loc.index)
       
     app= {"Code": loc,
           "Min": min_temp,
           "Max": max_temp,
           "Avg": avg_temp}
     new = new._append(app, ignore_index= True)
new
```

```python
plt.figure(figsize = (20,8))
plt.plot(new["Code"], new["Min"], color="b", label= 'Min Temp',marker= "o",alpha=overlapping)
plt.plot(new["Code"], new["Max"],color="r", label = "Max Temp",marker= "o",alpha=overlapping)
plt.plot(new["Code"], new["Avg"],color="g", label="Average Temp",marker= "o",alpha=overlapping)
plt.legend()
plt.xlabel('Station Codes')
plt.ylabel('Temperature (F)')
plt.title('Comparison of Temperatures')

```
